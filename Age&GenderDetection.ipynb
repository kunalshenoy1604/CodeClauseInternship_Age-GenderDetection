```python
import os
import cv2
import pandas as pd
from deepface import DeepFace

data = {
    "Name": [],
    "Age": [],
    "Gender": [],
    "Race": [],
    "Emotion": []
}

for file in os.listdir("People"):
    img_path = os.path.join("People", file)
    img = cv2.imread(img_path)
    
    if img is not None:
        results_all = DeepFace.analyze(img, actions=("age", "gender", "race", "emotion"))
        data["Name"].append(file.split(".")[0])
        data["Age"].append(results_all[0]["age"])
        data["Gender"].append(results_all[0]["dominant_gender"])
        data["Race"].append(results_all[0]["dominant_race"])
        data["Emotion"].append(results_all[0]["dominant_emotion"])
    else:
        data["Name"].append(file.split(".")[0])
        data["Age"].append(None)
        data["Gender"].append(None)
        data["Race"].append(None)
        data["Emotion"].append(None)

df = pd.DataFrame(data)
df.to_csv('dataset.csv', index=False)
print(df)



```
